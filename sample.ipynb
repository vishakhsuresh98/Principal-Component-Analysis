```python
# X = arr.flatten()
X = np.zeros([3, arr.shape[0]*arr.shape[1]])
```

```python
c = 0
for i in range(0, arr.shape[0]):
    for j in range(0, arr.shape[1]):
        X[0][c] = arr[i,j]
        X[1][c] = arr2[i,j]
        X[2][c] = arr3[i,j]
```

```python
XT = X.T
```

```python
U, s, V = np.linalg.svd(X)
```

```python
U.shape
```

```python
s.shape
```

```python
V.shape
```

```python
S = np.zeros((X.shape[0], X.shape[1]))
```

```python
S[:X.shape[1], :X.shape[1]] = np.diag(s)
```

```python
n_component = 2
S = S[:, :n_component]
V = V[:n_component, :]
B = U.dot(S.dot(V))
print(B)
```
